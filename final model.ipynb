```python
 data_no_duplicates['without_https']=data_no_duplicates['content'].str.replace('http[s]?://\S+','') # removing html content
```

```python
data_no_duplicates['content_new'] = data_no_duplicates['without_https'].str.replace(r'[^\w]+', ' ') # provided space between the words and special characters. Removed punctuations as well
                                                                  # also removed \n from the data     
```

```python
data_no_duplicates['content_without_no'] = data_no_duplicates['content_new'].str.replace(r'[\d]+', ' ') # removed numbers from the dataset
```

```python
data_new=data_no_duplicates[['Class','content_without_no']]
```

```python
data_new.head()
```

```python
data_new.isnull().sum() # check for missing values in each column
```

```python
import nltk
stopwords=nltk.corpus.stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer
wn=nltk.WordNetLemmatizer()
```

```python
dir(stopwords)
```

```python
def process(text):
    no_punc=''.join([char for char in text if char not in string.punctuation]).lower()
    clean_words=[wn.lemmatize(word) for word in no_punc.split() if word not in stopwords]
    return clean_words
    
```

```python
# without tfidf
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Convert a collection of text documents to a matrix of token counts
cv=CountVectorizer(analyzer=process)
```

```python
x=cv.fit_transform(data_new['content_without_no'])
```

```python
cv=CountVectorizer(analyzer=process)
```

```python
x=cv.fit_transform(data_new['content_without_no'])
```

```python
x.shape
```

```python
# Building multinomial naive bayes model
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,data_new['Class'],test_size=0.20,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE # importing smote for oversampling to counter imbalanced dataset
```

```python
smote=SMOTE()
```

```python
x_train_smote,y_train_smote=smote.fit_sample(x_train,y_train)
```

```python
from collections import Counter
```

```python
print("Before smote:", Counter(y_train))
print("After smote:", Counter(y_train_smote)) *# Checking whether over sampling worked properly
```

```python
classifier=MB().fit(x_train_smote,y_train_smote)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred=classifier.predict(x_train_smote)
print(classification_report(y_train_smote,pred))
```

```python
print(confusion_matrix(y_train_smote,pred))
```

```python
print('accuracy:', accuracy_score(y_train_smote,pred)) # accuracy of 0.93
```

```python
pred_test=classifier.predict(x_test)
```

```python
print(confusion_matrix(y_test,pred_test))
```

```python
print('accuracy:', accuracy_score(y_test,pred_test)) # 0.95
```
